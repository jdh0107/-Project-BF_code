##### crawl파일에 (괄호안)지우기

```python
import re
```

```python
nobracket=[]
for i in range(0,len(crawl['행정구역'])):
    if type(crawl['행정구역'][i])==str:
        nobracket.append(re.sub(r'\([^)]*\)','',crawl['행정구역'][i]).strip())
crawl['행정구역']=nobracket
```

```python
crawl[0:3]
```

##### 매출액과 crawl에서 겹치는 행정구역은 1181개

```python
num=0
for i in range(len(crawl['행정구역'])):
    for j in range(len(sales['상권_코드_명'])):
        if sales['상권_코드_명'][j] == crawl['행정구역'][i]:
            num+=1
num           
```

## 병합 시작

###### cs=crawl&sales

```python
cs=pd.merge(crawl,sales[['상권_코드_명','점포수','평균매출액','남성매출비율']],left_on='행정구역',right_on='상권_코드_명',how='left')
```

```python
cs[0:3]
```

##### csl=crawl&sales&live

```python
csl=pd.merge(cs,live.iloc[:,3:],left_on='행정구역',right_on='상권 코드 명',how='left')
```

```python
csl[0:3]
```

##### cslf=crawl&sales&live&facility

```python
cslf=pd.merge(csl,facility.iloc[:,4:],left_on='행정구역',right_on='상권_코드_명',how='left')
```

```python
cslf[0:3]
```

##### cslfh=crawl&sales&live&facility&house

```python
cslfh=pd.merge(cslf,house.iloc[:,4:7],left_on='행정구역',right_on='상권_코드_명',how='left')
```

```python
cslfh[0:3]
```

##### cslfho=crawl&sales&live&facility&house&office

```python
cslfho=pd.merge(cslfh,office.iloc[:,3:],left_on='행정구역',right_on='상권_코드_명',how='left')
```
